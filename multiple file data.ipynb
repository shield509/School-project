```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Specify the column for which you want to count occurrences
column_name = 'Product'

# Count occurrences of values in the specified column
value_counts = combined_df[column_name].value_counts().reset_index()
value_counts.columns = ['Product', 'Count']

# Plot the result using Seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x='Product', y='Count', data=value_counts, palette='viridis', edgecolor='black')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add gridlines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Labels and title
plt.xlabel('Product')
plt.ylabel('Count')
plt.title('Product Count')

# Show the plot
plt.tight_layout()
plt.show()
```

```python

# Specify the column for which you want to count occurrences
column_name = 'Product'

# Count occurrences of values in the specified column
value_counts = combined_df[column_name].value_counts()

# Plot the result using Matplotlib
plt.figure(figsize=(8, 8))
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(value_counts))))

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')  

# Title
plt.title('Product Distribution')

# Show the plot
plt.show()
```

```python


# Specify the column for which you want to count occurrences
column_name = 'Product'

# Count occurrences of values in the specified column
value_counts = combined_df[column_name].value_counts()

# Define colors and explode
colors = plt.cm.Paired(range(len(value_counts)))
explode = [0.1 if i == value_counts.argmax() else 0 for i in range(len(value_counts))]  # Highlight the largest slice

# Plot the result using Matplotlib
plt.figure(figsize=(15, 10))
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode, shadow=True)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')  

# Title
plt.title('Product Distribution')

# Show the plot
plt.show()
```
